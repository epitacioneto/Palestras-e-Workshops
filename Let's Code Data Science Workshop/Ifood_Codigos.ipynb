Finalizando a EDA - exploratory data analysis

```python
df.head()
```

```python
cols = ['Income', 'Recency', 'MntWines',
        'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
        'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
        'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Days_customer']
```

```python
df_cluster = df[cols]
```

```python
df_cluster.plot(kind='box');
```

$$\frac{x - media}{std}$$

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_cluster_scaler = scaler.fit_transform(df_cluster)
```

```python
df_cluster_scaler = pd.DataFrame(df_cluster_scaler, columns=cols)
df_cluster_scaler.plot(kind='box', figsize=(14, 3), rot=45);
```

```python
df_cluster_scaler.shape
```

```python
sns.scatterplot(data=df_cluster_scaler, x='Income', y='Recency');
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit_predict(df_cluster_scaler)
```

```python
kmeans.inertia_
```

```python
from tqdm import tqdm
```

```python
ks_lst = range(2, 21)
inertias_lst = []

for i in tqdm(ks_lst):
    kmeans = KMeans(i)
    kmeans.fit(df_cluster_scaler)
    inertias_lst.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(14, 3))
plt.xticks(ks_lst)
plt.plot(ks_lst, inertias_lst);
```

```python
kmeans = KMeans(3)
```

```python
clusters = kmeans.fit_predict(df_cluster_scaler)
```

```python
clusters
```

```python
df['clusters'] = clusters
```

```python
df.head()
```

```python
sns.boxplot(data=df, x='clusters', y='Year_Birth');
```

```python
sns.histplot(data=df, x='Year_Birth', hue='clusters');
```

```python
sns.displot(data=df, x='Year_Birth', hue='clusters', kind='kde');
```

```python
sns.displot(data=df, x='Income', hue='clusters', kind='kde');
```
