```python
cols_num = ['Year_Birth', 'Income', 'Recency', 'MntWines',
            'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
            'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
            'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Days_customer']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from tqdm import tqdm
```

```python
ks_lst = range(2, 21)
inertias_lst = []
```

```python
df_num = df[cols_num]
```

```python
for k in tqdm(ks_lst):
    pipe = Pipeline([('scaler', StandardScaler()),
                     ('kmeans', KMeans(n_clusters=k))])
    pipe.fit(df_num)
    inertias_lst.append(pipe['kmeans'].inertia_)
```

```python
plt.xticks(ks_lst)
plt.plot(ks_lst, inertias_lst);
```

```python
pipe = Pipeline([('scaler', StandardScaler()),
                 ('kmeans', KMeans(n_clusters=3))])

clusters = pipe.fit_predict(df_num)
```

```python
df_num['clusters'] = clusters
```

```python
df_num.head()
```

```python
df_num.columns
```

```python
_, ax = plt.subplots(2, 6, sharex=True, figsize=(18, 6))

sns.boxplot(data=df_num, x='clusters', y='MntWines', ax=ax[0, 0])
sns.boxplot(data=df_num, x='clusters', y='MntFruits', ax=ax[0, 1])
sns.boxplot(data=df_num, x='clusters', y='MntMeatProducts', ax=ax[0, 2])
sns.boxplot(data=df_num, x='clusters', y='MntFishProducts', ax=ax[0, 3])
sns.boxplot(data=df_num, x='clusters', y='MntSweetProducts', ax=ax[0, 4])
sns.boxplot(data=df_num, x='clusters', y='MntGoldProds', ax=ax[0, 5])

sns.boxplot(data=df_num, x='clusters', y='NumDealsPurchases', ax=ax[1, 0])
sns.boxplot(data=df_num, x='clusters', y='NumWebPurchases', ax=ax[1, 1])
sns.boxplot(data=df_num, x='clusters', y='NumCatalogPurchases', ax=ax[1, 2])
sns.boxplot(data=df_num, x='clusters', y='NumStorePurchases', ax=ax[1, 3])
sns.boxplot(data=df_num, x='clusters', y='NumWebVisitsMonth', ax=ax[1, 4])
sns.boxplot(data=df_num, x='clusters', y='Income', ax=ax[1, 5])

plt.tight_layout()
plt.show()
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
x = df.drop(columns=['Response'])
```

```python
cols_cat = ['Education', 'Marital_Status', 'Kidhome', 'Teenhome',
            'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
            'AcceptedCmp2', 'Complain']

cols_num = [i for i in x.columns if i not in cols_cat]
```

```python
scaler = StandardScaler()
```

```python
x_scaler = scaler.fit_transform(x[cols_num])
```

```python
x_scaler = pd.DataFrame(x_scaler, columns=cols_num, index=x.index)
```

```python
x_scaler = pd.concat([x_scaler, x[cols_cat]], axis=1)
```

```python
cols_cat_index = [x_scaler.columns.get_loc(c) for c in cols_cat if c in df]
```

```python
cost_lst = []

for k in tqdm(ks_lst):
    kprot = KPrototypes(n_clusters=k, random_state=42, n_jobs=-1)
    kprot.fit(x_scaler, categorical=cols_cat_index)
    cost_lst.append(kprot.cost_)
```

```python
plt.xticks(ks_lst)
plt.plot(ks_lst, cost_lst);
```

```python
kprot = KPrototypes(n_clusters=3, random_state=42, n_jobs=-1)
clusters = kprot.fit_predict(x_scaler, categorical=cols_cat_index)
```

```python
x['clusters'] = clusters
```
