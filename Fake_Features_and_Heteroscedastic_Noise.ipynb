```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_pts = 100
homoscedastic_noise_frac = 0.05
heteroskedastic_noise_frac = 0.25

X = np.random.uniform(1, 101, num_pts)
Y = 2 * X

# Add fake homoscedastic noise to our label
Y_max = np.amax(Y)
Y_basic_noise = Y_max * homoscedastic_noise_frac
Ybn = Y + np.random.uniform(-Y_basic_noise, Y_basic_noise, num_pts)

# Create fake heteroscedastic noise scaler as a function of X
# We are BASICALLY creating an envelope for our noise
heteroskedastic_slope = (heteroskedastic_noise_frac - homoscedastic_noise_frac)/(np.max(X) - np.min(X))
heteroskedastic_noise_per_X = (heteroskedastic_slope * X + homoscedastic_noise_frac) * np.random.uniform(-Y_max, Y_max, num_pts)
Yhs = Y + heteroskedastic_noise_per_X
```

```python
# Plotting
plt.figure(figsize=(12,6))
plt.title('Heteroscedastic Noise')
plt.scatter(X, heteroskedastic_noise_per_X)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('No, Homoscedastic, And Heteroscedastic Noise')
plt.plot(X, Y, color="black")
plt.scatter(X, Ybn, color="blue")
plt.scatter(X, Yhs, color="red")
plt.show()
```
